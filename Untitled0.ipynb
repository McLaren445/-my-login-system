```python
# Import the necessary tools from the STIX library
from stix2.v21 import (GranularMarking, Identity, Indicator, Bundle, TLP_RED, TLP_AMBER, TLP_GREEN)

# 1. Define the Granular Markings (Which color applies to which field?)
granular_red = GranularMarking(
    marking_ref=TLP_RED.id,
    selectors=["description"]
)

granular_green = GranularMarking(
    marking_ref=TLP_GREEN.id,
    selectors=["name", "pattern"]
)

# 2. Create the Identity
identity = Identity(
    name="Gotham National Bank",
    identity_class="organization"
)

# 3. Create the Indicator with a VALID STIX pattern
# The pattern uses a raw string (r"...") to handle backslashes for regex,
# and internal double quotes (\"...\") for STIX pattern string literals.
indicator = Indicator(
    name="Fake email address",
    description="Known to be used by threat actor",
    indicator_types=["malicious-activity"],
    pattern=r"[email-message:from_ref.value MATCHES \".*\\banking@g0thamnatl\\.com$\"]",
    pattern_type="stix",
    valid_from="2024-01-01T00:00:00Z",
    granular_markings=[granular_red, granular_green],
    created_by_ref=identity.id
)

# 4. Create the Bundle
bundle = Bundle(objects=[identity, indicator])

# 5. Print the result
print(bundle.serialize(pretty=True))
```

```python
# Import all required STIX components
from stix2.v21 import (
    GranularMarking,
    Identity,
    Indicator,
    ThreatActor,
    Relationship,
    Bundle,
    TLP_RED,
    TLP_AMBER,
    TLP_GREEN
)

print("=" * 50)
print("STIX GRANULAR MARKINGS EXAMPLE")
print("=" * 50)

# =============================================
# PART 1: Create TLP Marking Definitions
# =============================================
print("\n1. Creating TLP Marking Definitions...")

# TLP_RED, TLP_AMBER, and TLP_GREEN are built into the stix2 library
# They automatically have the correct IDs and structure

print(f"   TLP_RED ID: {TLP_RED.id}")
print(f"   TLP_AMBER ID: {TLP_AMBER.id}")
print(f"   TLP_GREEN ID: {TLP_GREEN.id}")

# =============================================
# PART 2: Create Granular Markings
# =============================================
print("\n2. Creating Granular Markings...")

# Apply TLP_RED to the 'description' field
granular_red = GranularMarking(
    marking_ref=TLP_RED.id,
    selectors=["description"]
)
print("   ✓ TLP_RED → description field")

# Apply TLP_AMBER to the second item in indicator_types array (index 1)
granular_amber = GranularMarking(
    marking_ref=TLP_AMBER.id,
    selectors=["indicator_types.[1]"]
)
print("   ✓ TLP_AMBER → indicator_types[1] (attribution)")

# Apply TLP_GREEN to multiple fields
granular_green = GranularMarking(
    marking_ref=TLP_GREEN.id,
    selectors=["indicator_types.[0]", "name", "pattern"]
)
print("   ✓ TLP_GREEN → indicator_types[0], name, pattern")

# =============================================
# PART 3: Create Identity Object
# =============================================
print("\n3. Creating Identity SDO...")

identity = Identity(
    name="Gotham National Bank",
    identity_class="organization",
    sectors=["financial-services"],
    contact_information="contact@gothamnational.com"
)
print(f"   Identity created: {identity.name}")
print(f"   ID: {identity.id}")

# =============================================
# PART 4: Create Indicator Object with Granular Markings
# =============================================
print("\n4. Creating Indicator SDO with Granular Markings...")

indicator = Indicator(
    name="Fake email address",
    description="Known to be used by The Joker",
    indicator_types=["malicious-activity", "attribution"],
    pattern="[email-message:from_ref.value MATCHES '.*\\\\banking@g0thamnatl\\\\.com$']",
    pattern_type="stix",
    valid_from="2024-01-01T00:00:00Z",
    granular_markings=[granular_red, granular_amber, granular_green],
    created_by_ref=identity.id
)
print(f"   Indicator created: {indicator.name}")
print(f"   Indicator types: {indicator.indicator_types}")
print(f"   Pattern: {indicator.pattern}")
print(f"   Granular markings applied: {len(indicator.granular_markings)} markings")

# =============================================
# PART 5: Create Threat Actor Object
# =============================================
print("\n5. Creating Threat Actor SDO...")

threat_actor = ThreatActor(
    name="The Joker",
    threat_actor_types=["terrorist", "criminal"],
    aliases=["Joe Kerr", "The Clown Prince of Crime"],
    roles=["director"],
    resource_level="team",
    primary_motivation="personal-satisfaction",
    object_marking_refs=[TLP_RED],  # Entire object is TLP_RED
    created_by_ref=identity.id
)
print(f"   Threat Actor created: {threat_actor.name}")
print(f"   Object marking (entire object): TLP_RED")

# =============================================
# PART 6: Create Relationship Object
# =============================================
print("\n6. Creating Relationship SRO...")

relationship = Relationship(
    relationship_type="indicates",
    source_ref=indicator.id,
    target_ref=threat_actor.id,
    object_marking_refs=[TLP_RED],  # Entire relationship is TLP_RED
    created_by_ref=identity.id
)
print(f"   Relationship created: {relationship.relationship_type}")
print(f"   Source: {relationship.source_ref}")
print(f"   Target: {relationship.target_ref}")

# =============================================
# PART 7: Create and Display Bundle
# =============================================
print("\n7. Creating STIX Bundle...")

bundle = Bundle(objects=[identity, indicator, threat_actor, relationship])

print("\n" + "=" * 50)
print("FINAL STIX BUNDLE (JSON OUTPUT)")
print("=" * 50)
print(bundle.serialize(pretty=True))

print("\n" + "=" * 50)
print("✅ SUCCESS! No errors detected.")
print("=" * 50)
```
